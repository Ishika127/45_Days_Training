```python
class Store:
    def __init__(self, Store_Id, Store_Name, Location):
        self.Store_Id = Store_Id
        self.Store_Name = Store_Name
        self.Location = Location
class Product(Store):
    def __init__(self, Store_Id, Store_Name, Location,
                 P_Id, P_Name, Price):
        super().__init__(Store_Id, Store_Name, Location)
        self.P_Id = P_Id
        self.P_Name = P_Name
        self.Price = Price
    def Disp(self):
        print("Store ID:", self.Store_Id)
        print("Store Name:", self.Store_Name)
        print("Location:", self.Location)
        print("Product ID:", self.P_Id)
        print("Product Name:", self.P_Name)
        print("Price:", self.Price)
class Employ(Store):
    def __init__(self, Store_Id, Store_Name, Location,
                 E_Id, E_Name, Salary):
        super().__init__(Store_Id, Store_Name, Location)
        self.__E_Id = E_Id
        self.E_Name = E_Name
        self.__Salary = Salary
    def Disp(self):
        print("Store ID:", self.Store_Id)
        print("Store Name:", self.Store_Name)
        print("Location:", self.Location)
        print("Employee ID:", self.__E_Id)
        print("Employee Name:", self.E_Name)
        print("Employee Salary:", self.__Salary)
E1 = Employ(101, "Reliance Mart", "Delhi",501, "Rahul", 35000)
P1 = Product(101, "Reliance Mart", "Delhi",1001, "Laptop", 55000)
print("Detail of E1 Employ:")
E1.Disp()
print()
print("Detail of P1 Product:")
P1.Disp()
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
        print("Vehicle Constructor")

class Car(Vehicle):
    def __init__(self, brand, doors):
        super().__init__(brand)   # Calling parent constructor
        self.doors = doors
        print("Car Constructor")

c = Car("Toyota", 4)

print("Brand:", c.brand)
print("Doors:", c.doors)
```
