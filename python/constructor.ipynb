```python
#multi- level  
class GrandParent:
    def __init__(self):
        print("GrandParent Constructor")

class Parent(GrandParent):
    def __init__(self):
        super().__init__()
        print("Parent Constructor")

class Child(Parent):
    def __init__(self):
        super().__init__()
        print("Child Constructor")

obj = Child()
```

```python
class GrandParent:
    def __init__(self):
        print("GrandParent Constructor")

class Parent(GrandParent):
    def __init__(self):
        GrandParent.__init__(self)
        print("Parent Constructor")

class Child(Parent):
    def __init__(self):
        Parent.__init__(self)
        print("Child Constructor")

obj = Child()
```

```python
class Vehicle:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

class Car(Vehicle):
    def __init__(self, brand, price, doors, fuel_type):
        Vehicle.__init__(self, brand, price)
        self.doors = doors
        self.fuel_type = fuel_type

class Bike(Car):
    def __init__(self, brand, price, doors, fuel_type, bike_type, mileage):
        Car.__init__(self, brand, price, doors, fuel_type)
        self.bike_type = bike_type
        self.mileage = mileage

    def display(self):
        print("Brand :", self.brand)
        print("Price :", self.price)
        print("Doors :", self.doors)
        print("Fuel Type :", self.fuel_type)
        print("Bike Type :", self.bike_type)
        print("Mileage :", self.mileage)

obj = Bike("Honda", 150000, 4, "Petrol", "Sports", 45)

obj.display()
```

```python
class Vehicle:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

class Car(Vehicle):
    def __init__(self, brand, price, doors, fuel_type):
        Vehicle.__init__(self, brand, price)
        self.doors = doors
        self.fuel_type = fuel_type

class Bike(Car):
    def __init__(self, brand, price, doors, fuel_type, bike_type, mileage):
        Car.__init__(self, brand, price, doors, fuel_type)
        self.bike_type = bike_type
        self.mileage = mileage

    def display(self):
        print("\nVehicle Details")
        print("Brand :", self.brand)
        print("Price :", self.price)
        print("Doors :", self.doors)
        print("Fuel Type :", self.fuel_type)
        print("Bike Type :", self.bike_type)
        print("Mileage :", self.mileage)

# User Input
brand = input("Enter Brand: ")
price = int(input("Enter Price: "))
doors = int(input("Enter Number of Doors: "))
fuel_type = input("Enter Fuel Type: ")
bike_type = input("Enter Bike Type: ")
mileage = int(input("Enter Mileage: "))

obj = Bike(brand, price, doors, fuel_type, bike_type, mileage)

obj.display()
```

```python
class User:
    def __init__(self, name, email):
        self.name = name
        self.email = email

class Customer(User):
    def __init__(self, name, email, address):
        User.__init__(self, name, email)
        self.address = address

    def display(self):
        print("\nCustomer Details")
        print("Name:", self.name)
        print("Email:", self.email)
        print("Address:", self.address)

class Seller(User):
    def __init__(self, name, email, shop_name):
        User.__init__(self, name, email)
        self.shop_name = shop_name

    def display(self):
        print("\nSeller Details")
        print("Name:", self.name)
        print("Email:", self.email)
        print("Shop Name:", self.shop_name)

class Admin(User):
    def __init__(self, name, email):
        User.__init__(self, name, email)

    def display(self):
        print("\nAdmin Details")
        print("Name:", self.name)
        print("Email:", self.email)


print("1. Customer")
print("2. Seller")
print("3. Admin")

choice = int(input("Enter Choice: "))

if choice == 1:
    name = input("Enter Name: ")
    email = input("Enter Email: ")
    address = input("Enter Address: ")

    c = Customer(name, email, address)
    c.display()

elif choice == 2:
    name = input("Enter Name: ")
    email = input("Enter Email: ")
    shop_name = input("Enter Shop Name: ")

    s = Seller(name, email, shop_name)
    s.display()

elif choice == 3:
    name = input("Enter Name: ")
    email = input("Enter Email: ")

    a = Admin(name, email)
    a.display()

else:
    print("Invalid Choice")
```
