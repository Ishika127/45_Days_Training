```python
def discount(amount):
    
   if amount >= 2000:
    discount = amount * 0.10
    final_amount = amount - discount
    print("You got a 10% discount!")
    print("Discount =", discount)
    print("Final amount =", final_amount)
   else:  
    print("No discount available.")
    print("Amount to pay =", amount)

discount(27834)
```
