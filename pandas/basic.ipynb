```python
#dictionary 
dt ={
    "Name":["Ishika","Priya","Prachi"],
    "Age":[19,18,19],
    "Rollno":[34,55,54]
    }
print(dt)
ds=pd.DataFrame(dt)
print(ds)
print('\t')
dv=pd.Series()
print(dv)
print('\t')


print(ds["Name"]) #acess by using indexing 
print(ds[["Name","Age"]]) #acess two together 
print(ds[["Name","Age","Rollno"]])
```

```python
#Indexing of row by iloc
# name of row by using loc 
print(ds.iloc[0])
print(ds.loc[1])
```

Data Manipulation  add column , remove column , update column

```python
ds["Grade"]=["A+","A+","A++"]
print(ds)
ds["Marks"]=[29,30,54]
print(ds)
```

```python
ds=ds.rename(columns={"Marks":"Rollno"})
print(ds)
```

```python
ds=ds.drop("Mraks",axis=1)
print(ds)
ds=ds.drop("Marks",axis=1)
```

```python
ds=ds.drop("Rollno",axis=1)
print(ds)
```

```python
ds["Rollno"]=[34,55,54]
print(ds)
```

```python
ds["Marks"]=[55,60,23]
print(ds)
```

Using .apply method we can used any specfic  and lambda function in pandas

```python
ds["Marks"] = ds["Marks"].apply(
    lambda x: x + 5
)

print(ds)
```
