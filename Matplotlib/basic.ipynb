```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [100,150,120,180,200]

plt.plot(
    x,
    y,
    color="blue",
    linewidth=3,
    linestyle="--",
    marker="o",
    markersize=8,
    markerfacecolor="yellow",
    markeredgecolor="black",
    label="Sales"
)

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()

plt.show()
```

sub plots : 
fig.ax=figure  combination of multiple graphs at once("bar,pie,box_line)

```python
fig,ax=plt.subplots(2,1)
ax[0].scatter(df["Pclass"],df["PassengerId"])
ax[1].bar(df["Pclass"],df["PassengerId"])

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))

ax[0].scatter(df["Pclass"], df["PassengerId"])
ax[1].bar(df["Pclass"], df["PassengerId"])

plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(6,8))

ax[0].scatter(df["Pclass"], df["PassengerId"])

ax[1].bar(df["Pclass"], df["PassengerId"])

plt.tight_layout()
plt.show()
```
